# EDA(Exploratory Data Analysis)

## Univariate Analaysis

```python
# identifying the outlier
sns.boxplot(data=df,x="price")## as we can see that there is a oulier present which is "100000" and mostly price is lie btw 0 to 20000
```

```python
 df=df[df["price"]<1500]
sns.boxplot(data=df,x="price")
```

```python
# Price Distribution
plt.figure(figsize=(8,5))
sns.histplot(data=df,x='price',bins=100)
plt.title("Price Distribution")
plt.ylabel("Frequency")
```

 ### Analaysis of avegage price by neighbourhood_group


```python
plt.figure(figsize=(8, 5))

sns.barplot(
    data=df,
    x="neighbourhood_group",
    y="price",
    estimator="mean"   
)

plt.title("Average Price by Neighbourhood Group")
plt.xlabel("Neighbourhood Group")
plt.ylabel("Average Price")

plt.show()
```

# Bi Variate Analysis

```python
df.columns

```

## Price Comparison Across Neighbourhood Groups and Room Types


```python

sns.barplot( data=df,x="neighbourhood_group",y="price",hue="room_type")
plt.title("Price Comparison Across Neighbourhood Groups and Room Types")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(data=df,x="number_of_reviews",y="price",hue="neighbourhood_group")
plt.title("Relationship Between Price and Recent Reviews by Neighbourhood Group")
```

## Pairwise Relationships Between Price, Availability, and Reviews by Room Type

```python
sns.pairplot(data=df,vars=['price','minimum_nights','number_of_reviews','availability_365'],hue="room_type")
plt.suptitle(
    "Pairwise Relationships Between Price, Availability, and Reviews by Room Type",
    y=1.02
)
```

## Geographical Representation Of Airbnb

```python
plt.figure(figsize=(10,8))
sns.scatterplot(data=df,x="longitude",y="latitude",hue="room_type")
```

```python
# heat map - correlation of one variable with others for numerical column

corr = df[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds']].corr()
corr

plt.figure(figsize=(8, 6))
sns.heatmap(data=corr, annot=True)
```
